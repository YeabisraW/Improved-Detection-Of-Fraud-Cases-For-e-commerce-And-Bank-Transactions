```python
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from collections import Counter
import pandas as pd

# Load datasets
fraud = pd.read_csv("../data/raw/Fraud_Data.csv")
credit = pd.read_csv("../data/raw/creditcard.csv")

# ----------------------------
# Select numeric columns only for SMOTE
# ----------------------------
numeric_cols_fraud = ['age', 'purchase_value']  # Add other numeric features if available
X_fraud_num = fraud[numeric_cols_fraud]
y_fraud = fraud['class']

X_credit = credit.drop(columns=['Class'])
y_credit = credit['Class']

# Apply SMOTE
smote = SMOTE(random_state=42)

X_fraud_res, y_fraud_res = smote.fit_resample(X_fraud_num, y_fraud)
X_credit_res, y_credit_res = smote.fit_resample(X_credit, y_credit)

# ----------------------------
# Plot Figure 7: E-commerce class distribution after SMOTE
# ----------------------------
plt.figure(figsize=(6,4))
plt.bar(['Non-Fraud', 'Fraud'], [Counter(y_fraud_res)[0], Counter(y_fraud_res)[1]], color=['skyblue','salmon'])
plt.title("Figure 11: Class Distribution after SMOTE (E-commerce)")
plt.ylabel("Number of Transactions")
plt.show()

# ----------------------------
# Plot Figure 8: Credit card class distribution after SMOTE
# ----------------------------
plt.figure(figsize=(6,4))
plt.bar(['Non-Fraud', 'Fraud'], [Counter(y_credit_res)[0], Counter(y_credit_res)[1]], color=['skyblue','salmon'])
plt.title("Figure 12: Class Distribution after SMOTE (Credit Card)")
plt.ylabel("Number of Transactions")
plt.show()
```
